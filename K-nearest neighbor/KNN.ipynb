```python
import pandas as pd
import numpy as np
from scipy.stats import mode

fruits = pd.read_csv('fruit_data_with_colors.txt', delimiter ='\t')
fruits.head()
```

```python
X = fruits[['mass','width','height']]
Y = fruits['fruit_label']
# Y = fruits['fruit_subtype']

X_train = X[0:49]
X_test  = X[49:]

Y_train = Y[0:49]
Y_test  = Y[49:]
```

```python
def euclideanDistance(p1, p2):
    distance = np.sqrt(np.sum(p1-p2)**2)
    return distance
```

```python
def KNN(X_train, X_test, Y_train, Y_test, k):
    predictions = []
    for test_points in X_test.to_numpy():
        distances = []
        for i in range(len(X_train)):
            distances.append(euclideanDistance((np.array(X_train.iloc[i])), test_points))
        
        distance_data = pd.DataFrame(data = distances, columns=['distance'], index = Y_train.index)
#         print(distance_data)
        
        k_neighbors = distance_data.sort_values(by = ['distance']) [:k]
        """The [:k] will select upto k smallest neighbors"""
#         print(k_neighbors)
        
        labels = Y_train.loc[k_neighbors.index]  
#         print(labels)
        
        voting = mode(labels).mode[0]
                             
        predictions.append(voting)
    return predictions
```

```python
k = 5
test = KNN(X_train, X_test, Y_train, Y_test, k)
test
```

```python
# We can see that the majority of items in the list are 3. So the accuracy of the model is pretty low as it guessed
# all the fruit's in the test dataset incorrectly. In general, for almost any value of k it always gives inaccurate results
# the solution might be to randomly split the test and train to get better accuracy.
```
